```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph(comment='My DAG')

# Add nodes with custom names
dot.node('A', 'Start ')
dot.node('B', 'Generator')
dot.node('C', 'Evaluator')
dot.node('D', 'End ')

# Set the graph to use curved splines
  # Enable curved edges

# Add directed edges (A -> B -> C -> D)
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D')

# Now add a feedback loop from C -> B with a curve
dot.edge('C', 'B', xlabel='               if score < thres\nsend feedback', 
         constraint='false',)

# Render and show the graph directly in the notebook
dot.render('dag_output', format='png', cleanup=False)
from IPython.display import Image
Image(filename='dag_output.png')
```
